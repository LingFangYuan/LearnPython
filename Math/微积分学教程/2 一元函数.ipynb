### 第二章 一元函数
#### 一、函数的概念
##### 43、变量及其变动区域
- 数集 $X = \{x\}$ 给出变量 x 可能取的那些数值。X 内的每一个数值，x 都有机会能取到，这数集 X 就称为变量 x 的**变动区域**。一般地说，任一数集总可以当作变量的变动区域。
- 数的几何解释就是数轴上的点。变量 x 的变动区域 X 在这轴上就表示为某一点集。因此，变量的数值常称为点。
- 在分析上通常研究的是所谓的**连续的**或**密接的**变动着的变量。数的区间就是此类变量的变动区域。最常用的是有限区间，它用两个实数 a 及 b $(a < b)$端点为界限，有些端点可以包含在区间内，也可以不包含在内。
- **闭区间** [a,b]: $a \le x \le b$ （包含两个端点）；
- **半开区间** $\begin{cases} (a, b]: a < x \le b \\ [a, b): a \le x < b \end{cases}$ （仅包含一端点）；
- **开区间** (a, b): $a < x < b$ （不包含任一端点）。
- 在每一种情形都称数 $b - a$ 为**区间的长**。
- **无穷区间** ：用 $+\infty，-\infty$ 作为它的一端或两端。
 - $(-\infty, +\infty)$ : 全体实数集；
 - $(a, +\infty)$ : 满足不等式 $x > a$ 的数 x 所成的集；
 - $(-\infty, b]$ : 由不等式 $x \le b$ 所决定。
 - 无穷区间的几何表示，是两端无线伸展的直线或仅一端趋于无限的射线。

##### 44、变量间的函数关系，例题
- 数学分析内主要研究的事物，并非一个变量独自地变动，而是两个或多个变量在同时变动时相互之间的关系。在这里只限于研究两个变量的最简单情形。
- 当其中之一（**自变量**）已给定具体的数值时，则另一变量（**因变量**或**函数**）的数值也就确定了。
- 1) 圆的面积 Q 是它的半径 R 的函数；它的数值可以根据给定的半径数值用已知公式：$Q = πR^2$ 计算出来。
- 2) 自有落体不计阻力的情形。由运动开始时算起的时间 t(秒) 与在这时间内经过的路程 s(米)依靠方程：$s = \frac{gt^2}{2}$ 联结着，式中 $g = 9.81 米/秒^2$。路程 s 就成为经过时间 t 的函数。

##### 45、函数概念的定义
- **设给定两变量 x 及 y ，其变动区域为 X 及 Y 。假定根据问题的条件，变量 x 可以不受任何限制地取区域 X 内的任意数值。那么，如果依某一法则或规律，对于 X 中的每一 x 值总有一个确定的数值 y （在 Y 内）和它对应，则变量 y 就称为变量 x （在它的变动区域 X 内）的函数。**
- 自变量 x 亦称为函数的**变元**。
- 定义内存在两个要素：
  - 第一，指出变元 x 的变动区域 X（它称为函数的定义域）。
  - 第二，确定 x 与 y 的数值之间的对应法则或规律（函数 y 的变动区域 Y 通常并不指出，因为对应的规律本身就已经确定函数值的集合了）。
- 假设对应于 X 内的 x 的每一个数值，y 的数值不止一个，而是几个，甚至是无穷多个，在这种场合函数称为**多值的**，以区别于前面定义的**单值函数**。
- y 是 x 的函数这一事实，记成：$y = f(x)$ 。
- 变量的数值之间的对应法则或规律，是函数关系这概念的要素。这法则可以有各种各样的表现方式。
 - 解析表示法，它指示出，对 x 的数值及一些常数必须进行哪些演算，才可以得出 y 的对应的数值。
 - 列表法。
 - 图示法。

##### 46、函数的解析表示法
- $1^。$ 在这些公式内可以进行怎样的演算？
 - 第一，初等代数三角学内研究过的一切演算：算术运算、乘幂、取对数，由角度求三角函数值及其反运算。
 - 还需要增加其它的演算，首要的就是极限步骤。
- $2^。$ 须注意借解析式或公式所表示的函数的定义域。  
 - 每一个包含变元 x 的解析式具有所谓自然的适用区域：就是使这解析式子有意义的一切 x 所成的集合。解析式子有意义，既是指它有着完全确定的有限实数值的意思。  
 - 设在任何具体问题内，变量 x 由于事情的本质被限制在其变动区域 X 内，而此问题，引导我们去考察一个具有解析表达式的函数 $f(x)$ 。虽然这个表达式可能在区域 X 以外也有意义。但欲越出界限来研究具体的问题却是全然不可能的。在这里，解析式便只担任着附属的辅助角色了。
- $3^。$ 对于变元的一切数值函数并非有同一公式所确定，而是对于变元的某一部分数值用某一公式，对于另一部分数值用另一公式。
 - $f(x) = \begin{cases} 1: |x| > 1，\\ -1: |x| < 1， \\ 0: x = \pm 1 \end{cases}$
 - 狄利克雷(P.G.Lejeune-Dirichlet)函数。$X(x) = \begin{cases} 1: x是有理数，\\ 0: x是无理数 \end{cases}$
 - 克罗内克(L.Kronecker)函数，称它为 “x 的符号”并记成sgnx: $sgn(x) = \begin{cases} 1: x > 0，\\ -1: x < 0，\\ 0: x = 0 \end{cases}$
 - 不必以为对于全部 x 的数值用一个公式给定的函数与用几个公式来定义的函数之间有着原则上的差别，通常，用几个公式给定的函数也可以用一个公式给定它。  
 例如，运用求极限的演算，则上面引入的第一个函数 $f(x)$ 就可以用一个公式来给定它: $f(x) = \lim \frac{x^{2n} - 1}{x^{2n} + 1}$ 。  
 在 $|x| > 1$ 时，幂 $x^{2n} \to +\infty$ ，而它的倒数趋于 0 ，因此：$\lim \frac{x^{2n} - 1}{x^{2n} + 1} = \lim \frac{1 - \frac{1}{x^{2n}}}{1 + \frac{1}{x^{2n}}} = 1$ 。  
 在 $|x| < 1$ 时，幂 $x^{2n} \to 0$ ，在这种情形：$\lim \frac{x^{2n} - 1}{x^{2n} + 1} = -1$ 。  
 最后， 在 $x= \pm 1$ 时，显然 $x^{2n} = 1$ ，由此：$\lim \frac{x^{2n} - 1}{x^{2n} + 1} = 0$ 。  
 一切这些都符合原来的定义。

##### 49、反函数的概念
- 假定在某一区域 X 内给定函数 $y = f(x)$ ，并设当 x 在区域 X 内变动时，一切函数值所成的集是 Y 。由区域 Y 内选取任一数值 $y = y_0$ ，则在区域 X 内必能求出数值 $x = x_0$ ，使得函数在 $x_0$ 所取得数值刚好就是 $y_0$ ，即 $f(x_0) = y_0$ ，像这样的数值 $x_0$ 可能出现好多个。因此，Y 内的任一数值 y 将与一个或几个 x 的数值相对应，由此对应地确定在区域 Y 内的单值或多值函数 $x = g(y)$ ，它就称为函数 $y = f(x)$ 的**反函数**。
- 如果函数 f 的反函数是单值的，那么函数本身称为**可逆的**，通常谈到反函数时，无条件地假定它是单值的。
- 按照函数 $y = f(x)$ 的图像很容易判断它的反函数 $x = g(y)$ 是单值或不是单值的。若平行于 x 轴的直线与这图像最多相交于一个点，就出现第一种情形。反之，若这种直线中的某几条可与图像相交于几个点，则反函数就是多值的。

##### 51、函数的叠置
- 以另一变元的函数作为已给函数的变元，便形成**函数叠置**，例如，叠置函数 $\sin x$ 及 $\lg y$ 就得出函数 $\lg \sin x$ 。
- 假定函数 $y = f(x)$ 对于区域 X 内的 x 是有意义的，并且它的数值全部包含在区域 Y 内。再设函数 $z = φ（y）$ 刚好在区域 Y 内是有意义的，则常说，变量 z 借 y 做媒介而成为 x 的函数：$z = φ(f(x))$ ，所得的复合函数，就是叠置函数 $f(x)$ 及 $φ(y)$ 的结果。
- 一种函数关系须借复合函数来表出的这一特性，并不依赖于这函数关系的本身，而只依赖于表示这关系的方法。

#### 二、函数的极限
##### 52、函数的极限的定义
- 考察数集 $X = \{x\}$ 。若在点 a 的任意近处包含有 X 中异于 a 的 x 值，则点 a 称为这数集的**聚点**。
- 以点 a 为中心的开区间 (a-δ, a+δ) 就称为点 a 的**邻域**。若在点 a 的任一**邻域**内包含 X 中异于 a 的 x 值，则点 a 是数集 X 的**聚点**。
- 聚点本身可以属于 X 或不属于 X 。
- **设在区域 X 内给定函数 $f(x)$ ，且 a 是 X 的聚点。若对于任一数 $ε > 0$ 能求出数 $δ > 0$ ，只需 $|x - a| < δ$ ，能使 $|f(x) - A| < ε$ ，则称当 x 趋向于 a 时，函数 $f(x)$ 以数 A 为极限。这一事实记成：$\lim\limits_{x \rightarrow a} f(x) = A \ \ \ (2)$ 。**
- 设 X 是这样一种区域，仅在 a 的右边任意近处，能找到 X 内的异于 a 的 x 数（在这种场合点 a 称为 X 的右聚点），则可以把刚才所给函数的极限的定义特殊化，使仅限于 $x > a$ 的数值。在这种场合，若函数的极限存在，就称为当 x 从右边趋向于 a 时函数 $f(x)$ 的极限，或简称**右极限**，并记成：$\lim\limits_{x \rightarrow a+0} f(x)\ 或\ f(a + 0)$ 。  
类似地：左聚点，当 x 从左边趋向于 a 时函数的极限**左极限**：$\lim\limits_{x \rightarrow a-0} f(x)\ 或\ f(a - 0)$ 。
- 若点 a 同时为 X 的右聚点及左聚点，极限 $\lim\limits_{x \rightarrow a} f(x) = A$ 存在的必要且充分条件为右极限及左极限各自存在而且相等：$\lim\limits_{x \rightarrow a+0} f(x) = \lim\limits_{x \rightarrow a-0} f(x) = A$ 。
- 当 x 趋向于有限极限 a 时，函数亦可以有无穷极限。当 x 趋向于 a 时，函数 $f(x)$ 以 $+\infty(-\infty)$ 为极限。记成：$\lim\limits_{x \rightarrow a} f(x) = +\infty(-\infty)$ 。
- 若数集 X 包含（绝对值）任意大的正（负）值 x ，则称 $+\infty(-\infty)$ 是 X 的聚点。当 x 趋向于 $+\infty(-\infty)$ 时函数 $f(x)$ 有极限 A ，并写成：$\lim\limits_{\substack{x \rightarrow \infty \\ (x \rightarrow -\infty)}} f(x) = A (或+\infty、-\infty)$ 。
- 函数 $f(x)$ 可以任意地“接近”于其极限 A ，只需自变量 x 充分地“接近”于它自己的极限 a 。
- 当函数 $f(x)$ 趋向于 0 时，它称为**无穷小**，若 $f(x)$ 趋向于 $\infty$ ，它就称为**无穷大**。

##### 53、变成整序变量的情形。
- 若将整序变量视为在自然数序范围内变动着的自变量 n 的函数，则当 $n \rightarrow \infty$ 时这函数的极限，显然与整序变量的极限相一致。**整序变量的极限是函数的极限的特殊情形**。
- 在某些意义上，函数的极限可归结于整序变量的极限。
- 设数集 $X = \{x\}$ 有聚点 a （a 可以是有限数，也可以是不同符号无穷）。则由 X 内可以取出以 a 为极限的 x 的（异于 a 的）序列：
$$x_1, x_2, x_3, ...\ \ \ (6)$$
与变元的序列（6）对应着的是函数的序列：
$$f(x_1), f(x_2), f(x_3), ...\ \ \ (7)$$
在等式（2）成立时这序列恒收敛于 A 。
若给定任意数 $ε > 0$ ，则首先根据极限的定义（2），能取出与它对应的数 $δ > 0$ 。由于序列（6）收敛于 a ，所以根据数 δ，必能求出序号 N ，使 $n > N$ 时能成立不等式 $|x_n - a| < δ$ ，又有 $|f(x_n) - A| < ε$ 。由此亦证明序列（7）收敛于 A。
- 假设自变量 x 依着以 a 为极限的任意序列（6）递变时，与它对应的函数的序列（7）恒有极限 A 。则这数 A 就是函数 $f(x)$ 的极限。  
由反面推论，假定 A 并非函数的极限。那时对于某些数 $ε > 0$ ，就没有对应的 δ 存在；即不论取怎样小的 δ ，至少能求出变数 x 的一个数值 $x = x'（异于 a）$ ，虽然 $|x' - a| < δ$ ，但仍有 $|f(x') - A| \ge ε$ 。  
取一趋于零的正数 ${δ_n}$ 的序列。根据刚才所说的，对于任一数 $δ = δ_n$ 恒能找出数值 $x' = x_n'$ ，虽然 $|x_n' - a| < δ_n$ ，但仍有 $|f(x_n') - A| \ge ε$ 。  
于是这些数值便组成某一序列：
$$x_1', x_2', x_3', ..., x_n'...$$
对于它们恒有：$|x_n' - a| < δ_n(n = 1, 2, 3, 4,...)$ 因 $δ_n \rightarrow 0$ ，故 $x_n' \rightarrow a$ 。  
依定理的假定，对应的函数的序列：
$$f(x_1'), f(x_2'), ..., f(x_n'), ...$$
应趋于 A ，但这是不可能的，因对应一切的 $n = 1, 2, 3, ...$ 恒有 $f(x_n') - A| \ge ε$ 。所得的矛盾就证明了命题。
- 函数的极限的概念的第二定义：**对于任何以 a 为极限的序列（6），对应的序列（7）常有极限 A** 。
- 假定对应于任何收敛于 a 的序列（6），序列（7）的极限常存在，就足以推得一切这些极限是重合的。

##### 54、例题
- 1) 证明 $\lim\limits_{x \rightarrow +\infty} a^x = +\infty\ (a > 1)$ 。  
对于任何的 $E > 0$ ，只要 $ \Delta = \log_a E$ ，就可以由 $x > \Delta$ 导出 $a^x > E$ 。  
仿此可以证明 $\lim\limits_{x \rightarrow -\infty} a^x = 0\ (a > 1)$ 。  
即不论 $ε > 0$ 怎样小，若取 $\Delta = \log_a \frac{1}{ε} = - \log_a ε$ ，则在 $x < -\Delta$ 时，必有 $a^x < ε$ 。
- 2) 证明 在 $a > 1$ 时，$\lim\limits_{x \rightarrow +\infty} \log_a x = +\infty ，\lim\limits_{x \rightarrow +0} \log_a x = -\infty$ 。  
对于任何给定的 $E > 0$ ，只需 $x > a^E$ ，就有 $\log_a x > E$ ，因而仿此，只需 $0 < x < a^{-E}$ ，就成立不等式 $\log_a x < -E$ 。
- 3) 证明 $\lim\limits_{x \rightarrow +\infty} \arctan x = \frac{π}{2}\ ，\lim\limits_{x \rightarrow -\infty} \arctan x = -\frac{π}{2}$ 。  
对于任何 $ε > 0$ ，取 $x > \tan (\frac{π}{2} - ε)$ ，就可以使 $\arctan x > \frac{π}{2} - ε$ ，于是 $0 < \frac{π}{2} - \arctan x < ε$ 。  
仿此，取 $x < \tan (ε - \frac{π}{2})$ ，就可以使 $\arctan x < ε - \frac{π}{2}$ ，于是 $0 < \arctan x - (-\frac{π}{2}) < ε$ 。
- 4) 证明 $\lim\limits_{x \rightarrow +\infty} \frac{a^x}{x} = +\infty(a > 1)$ 。  
回忆已经遇到过它的特例: $\lim\limits_{n \rightarrow +\infty} \frac{a^n}{n} = +\infty$ ，同时又有 $\lim\limits_{n \rightarrow +\infty} \frac{a^n}{n+1} = +\infty$ 。  
因此，根据给定的 $E > 0$ 能求出自然数 N ，使在 $n > N$ 时，成立不等式 $\frac{a^n}{n+1} > E$ 。  
今设 $x > N + 1$ ；若假定 $n = E(x)$ ，则 $n > N\ 又\ n \le x < n + 1$ ，于是 $\frac{a^x}{x} > \frac{a^n}{n+1} > E$ ，这就证明了命题。  
由此，很易求得 $\lim\limits_{x \rightarrow +\infty} \frac{a^x}{x^k} = +\infty(a > 1，k > 0)$ 。
- 5) 类似地，根据以前的结果 $\lim\limits_{n \rightarrow +\infty} \frac{\log_a n}{n} = 0 (a > 1)$ ，一般可以建立$\lim\limits_{x \rightarrow +\infty} \frac{\log_a x}{x} = 0 (a > 1)$ ，式中 x 是正实数。  
把这里的 x 换成 $x^k(k > 0)$ 容易证明 $\lim\limits_{x \rightarrow +\infty} \frac{\log_a x}{x^k} = 0 (a > 1, k > 0)$ 。  
事实上，若对任意给定的 $ε > 0$ ，取这样的 &Delta; ，使当 $x > \Delta$ 时已能满足不等式 $\frac{\log_a x}{x} < kε$ ，则在 $x > \Delta_1 = \Delta^{\frac{1}{k}}$ 时就有 $x^k > \Delta$ ，因而 $\frac{\log_a x}{x^k} < ε$ 。  
若把这里的 x 换成 $\frac{1}{x}$ ，则已得的结果可以改写成 $\lim\limits_{x \rightarrow +0} x^k\log_a x = 0(a > 1, k> 0)$。
- 6) 由已证明的极限关系 $\lim\limits_{n \rightarrow +\infty} a^{\frac{1}{n}} = 1$ ，可以得出更普遍的极限关系 $\lim\limits_{x \rightarrow 0} a^x = 1$ 。  
显然有 $\lim\limits_{n \rightarrow +\infty} a^{-\frac{1}{n}} = \lim\limits_{n \rightarrow +\infty} \frac{1}{a^{\frac{1}{n}}} = 1$ 。  
因此，不论 $ε > 0$ 怎样，可以求出自然数 $n_0$ ，使（若 $a > 1$） $1 - ε < a^{-\frac{1}{n_0}} < a^{\frac{1}{n_0}} < 1 + ε$ ，  
今若 $-\frac{1}{n_0} < x < \frac{1}{n_0}$ ，则 $a^{-\frac{1}{n_0}} < a^x < a^{\frac{1}{n_0}}$ ，  
由此 $1 - ε < a^x < 1 + ε$ 或 $|a^x - 1| < ε$ 。这便证明了命题。
- 7) 现在将建立下面的结果： $$\lim\limits_{x \rightarrow 0} \frac{\sin x}{x} = 1 \ \ \ (8)$$
但是首先必须证明下列有用的不等式：
$$\sin x < x < \tan x \ (0< x <\frac{π}{2}) \ \ \ (9)$$
为这目的，考察在半径为 R 的圆内的锐角 &ang;AOB ，弦AB 及切圆于点 A 的切线 AC （图22）。![22.png](./images/22.png) 
则有：三角形 AOB 的面积 < 扇形 AOB 的面积 < 三角形 AOC 的面积。若用 x 表示角 &ang;AOC 的弧度，于是弧 $\mathop{AB}\limits^{\frown}$ 的长就可以用 Rx 来表示，而这些不等式就可以改写成：
$$\frac{1}{2}R^2 \cdot \sin x < \frac{1}{2}R^2 \cdot x < \frac{1}{2}R^2 \cdot \tan x$$
由此约去 $\frac{1}{2}R^2$ 就得出不等式 (9)。  
在假定 $0 < x < \frac{π}{2}$ 之下，用不等式（9）的各项去除 $\sin x$ ，则得 $1 > \frac{\sin x}{x} > \cos x$ 。  
由此 $0 < 1 - \frac{\sin x}{x} < 1 - \cos x$ 。  
但 $1 - \cos x = 2\sin^2 \frac{x}{2} < 2\sin \frac{x}{2} < x$ 。  
于是 $0 < 1 - \frac{\sin x}{x} < x$ 。由此推得 $|\frac{\sin x}{x} - 1| < |x|$ 。  
显然改变 x 的符号时，上式仍旧是正确的，就是说，对于一切 $x \ne 0$ ，只需 $|x| < \frac{π}{2}$ ，这不等式总成立。所得的不等式就证明了（8）式。  
事实上，对于任意给定的数 $ε > 0$ ，只需选取数 ε 及 $\frac{π}{2}$ 中的最小者作为 δ，在 $|x| < δ$ 时，首先这不等式是可用的（$δ \le \frac{π}{2}$），而用了它（因 $δ \le ε$），就得到 $|\frac{\sin x}{x} - 1| < ε$ 。  
依函数极限的定义，这就表示，函数 $\frac{\sin x}{x}$ 在 x 趋向于 0 时以 1 为极限，于是关系式（8）被证明了。
- 7a) 极限关系式（8）根据 53 可以理解为：只需 x 依收敛于零的序列 ${x_n}$ 而递变，整序变量 $\frac{\sin x}{x}$ 终是趋于 1 。应用这事实来求整序变量的极限 $\lim\limits_{x \rightarrow \infty} \cos \frac{\phi}{2} \cdot \cos \frac{\phi}{2^2} \cdot ... \cdot \cos \frac{\phi}{2^n}$ ，式中 &phi; 是任何异于 0 的数。  
显然：
$$\sin \phi = 2\cos \frac{\phi}{2} \cdot \sin \frac{\phi}{2} = 2^2\cos \frac{\phi}{2} \cdot \cos \frac{\phi}{2^2} \cdot \sin \frac{\phi}{2^2} = ... $$
$$= 2^n\cos \frac{\phi}{2} \cdot \cos \frac{\phi}{2^2} \cdot ... \cdot \cos \frac{\phi}{2^n} \cdot \sin \frac{\phi}{2^n}$$
于是式子的形状就表示为: 
$$\frac{\sin \phi}{\phi} \cdot \frac{\frac{\phi}{2^n}}{\sin \frac{\phi}{2^n}}$$
因为 $x_n = \frac{\phi}{2^n} \rightarrow 0$ ，故依上述的命题 $\lim \frac{\sin \frac{\phi}{2^n}}{\frac {\phi}{2^n}} = 1$ ，于是本题内的整序变量的极限就等于数 $\frac{\sin \phi}{\phi}$ 。
- 8) 再研究一个十分重要的极限。就是，在 36 内曾定义数 e 作为整序变量的极限：$$e = \lim\limits_{n \rightarrow +\infty} (1 + \frac{1}{n})^n \ \ \ (10)$$
今建立更普遍的结果 $$\lim\limits_{x \rightarrow +\infty} (1 + \frac{1}{x})^x = e \ \ \  (11)$$
及同样
$$\lim\limits_{x \rightarrow -\infty} (1 + \frac{1}{x})^x = e \ \ \  (11a)$$
应用借“序列的语言”来表达的极限的第二个定义。与（10）同时成立下面的等式 $$\lim\ (1 + \frac{1}{n_k})^{n_k} = e \ \ \  (12)$$
只要 {$n_k$} 是自然数的任意序列，随标号 k 一起增大以至无穷。  
今设 x 依任何趋向于 $+\infty$ 的序列 {$x_k$} 而递变，并且可以当做一切 $x_k > 1$ 。令 $n_k = E(x_k)$ ，于是 $n_k \le x_k < n_k + 1 ，又\ n_k \to +\infty$ 。  
因为这时 $\frac{1}{n_k + 1} < \frac{1}{x_k} \le \frac{1}{n_k}$ ，所以 $(1 + \frac{1}{n_k+1})^{n_k} < (1 + \frac{1}{x_k})^{x_k} < (1 + \frac{1}{n_k})^{n_k+1}$ 。  
两个靠边的式子可以改写成：
$$(1 + \frac{1}{n_k+1})^{n_k} = \frac{(1 + \frac{1}{n_k+1})^{n_k + 1}}{1 + \frac{1}{n_k+1}}$$
$$(1 + \frac{1}{n_k})^{n_k+1} = (1 + \frac{1}{n_k})^{n_k} \cdot (1 + \frac{1}{n_k})$$
根据（12），$(1 + \frac{1}{n_k})^{n_k} \to e$ ，同样地有 $(1 + \frac{1}{n_k+1})^{n_k+1} \to e$ ，  
同时显然有 $1 + \frac{1}{n_k} \to 1$ ，$1 + \frac{1}{n_k+1} \to 1$ ，  
这样上述的两式都趋于公共极限 e ，所以夹在他们中间的式子也应趋于 e ：$\lim\ (1 + \frac{1}{x_k})^{x_k} = e$ 。  
由此，关系式（11）内的第一已经在“序列的语言”下获得证明。  
为了再要证明（11a），可以假定序列 {$x_k$} 以 $-\infty$ 为极限（并且可以当作一切 $x_k < -1$）。若令 $x_k = - y_k$ ，则 $y_k \to +\infty\ (y_k > 1)$ 。显然 
$$(1 + \frac{1}{x_k})^{x_k} = (1 - \frac{1}{y_k})^{-y_k} = (\frac{y_k - 1}{y_k})^{-y_k} = (\frac{y_k}{y_k - 1})^{y_k} = (1 + \frac{1}{y_k - 1})^{y_k}$$
$$= (1 + \frac{1}{y_k - 1})^{y_k - 1} \cdot (1 + \frac{1}{y_k - 1})$$
因为，依前面所证明的，最后一式的第一因式趋向于 e ，第二因式显然以 1 为极限，所以左边的式子也应趋于 e 。故式（11a）已证实。  
今把表达式 $(1 + \frac{1}{x})^x$ 内的变量 x 换成 $\frac{1}{α}$ ，若 α 是趋向于 0 的序列中的一值，则 $x = \frac{1}{α}$ 将趋向于 $\pm\infty$ 。因此，公式（11）和（11a）可以改写成 
$$e = \lim\limits_{α \to 0} (1 + α)^{\frac{1}{α}}\ \ \ (13)$$
- 9) 函数 $\sin x$ 在 x 趋向于 $\pm\infty$ 时就完全没有极限。  
站在“序列的观点”上来说明极限的不存在最为简单，只要注意 x 值的两个序列
$$\{(2n - \frac{1}{2})π\}\ 及\ \{(2n + \frac{1}{2})π\}\ （n=1, 2, 3, ...）$$
都以 $+\infty$ 为极限，而与它们对应函数值的序列确趋于两个相异的极限：
$$\sin (2n - \frac{1}{2})π = -1 \to -1，\sin (2n + \frac{1}{2})π = 1 \to 1$$
又可以改用这样的说法：若取以 $+\infty$ 为极限的 x 值的序列
$$\{(n + \frac{1}{2})π\}\ （n = 1, 2, 3, ...）$$
则与它对应的函数值的序列：
$$\sin (n + \frac{1}{2})π = (-1)^n\ （n = 1, 2, 3, ...）$$
就全然没有极限。  
若回想起正弦曲线的“振动”的特性，则在所考察的情形极限并不存在是很明显的。  
类似地，函数 $\sin \frac{1}{α}$ 在 α 趋向于 0 时极限并不存在。  
在表达式 $\sin \frac{1}{α}$ 内，将字母 α 仍旧写成字母 x ，并考察当 x 的数值在由 0 至 $\frac{2}{π}$（及由 $-\frac{2}{π}$ 至 0）的范围内时函数 $y = \sin \frac{1}{x} （x \ne 0）$ 的图像。  
依次减小到 0 的 x 数值：
$$\frac{2}{π}, \frac{1}{π}, \frac{2}{3π}, \frac{1}{2π}, \frac{2}{5π}, ..., \frac{2}{(2n-1)π}, \frac{1}{nπ}, ...$$
与它们对应的是增大至无穷大的 $\frac{1}{π}$ 数值：
$$\frac{π}{2}, π, \frac{3π}{2}, 2π, \frac{5π}{2}, ..., \frac{(2n-1)π}{2}, nπ, ...$$
在上诉当 x 减小时各相邻数值之间的诸区间内，函数 $\sin \frac{1}{x}$ 更迭地由 1 减小至 0 ，再由 0 减小至 -1 ，然后又由 -1 增大至 0 ，再由 0 增大至 1。  
这样，函数 $\sin \frac{1}{x}$ ，类似于函数 $\sin x$ ，就有着无数次的振动，但 $\sin x$ 的振动散布于无穷区间，而此处的 $\sin \frac{1}{x}$ 的振动却见于有限区间中，且凝聚于 0 。（图 23）![23.png](./images/23.png)
- 10) 若考察函数 $x\sin \frac{1}{x}(在 x \ne 0 时)$ ，它与刚才研究过的函数 $\sin \frac{1}{x}$ 仅相差一个乘数 x ，则这次在 $x \to 0$ 时就有极限存在了：  $\lim\limits_{x \to 0} x \cdot \sin \frac{1}{x} = 0$ ，这事实由不等式 $|x \cdot \sin \frac{1}{x}| \le |x|$ 就可以立刻明白。  
在 x 接近于 0 时，我们的函数仍旧发生无数次的振动，但它们的振幅（由于乘数 x）渐渐减小而趋于 0 。由此就保证了极限的存在。函数 $y = x \cdot \sin \frac{1}{x}$ 的图像画在 图24 内：它包藏在两条坐标角的分角线 $y = x$ 与 $y = -x$ 之间。![24.png](./images/24.png)  
- **附注** 我们已有一系列的极限 $\lim\limits_{x \to 0} \frac{\sin x}{x} = 1，\lim\limits_{x \to 0} (1 + x)^{\frac{1}{x}} = e，\lim\limits_{x \to 0} x \cdot \sin \frac{1}{x} = 0$ ，具有一种共同的特性：即此处所考察的诸函数中没有一个函数在 $x = 0$ 时是有定义的。但这并不影响我们说及当 $x \to 0$ 时它们的极限的存在，因为，依照 52 中所定义的准确定义，刚好数值 $x = 0$ 是不被考察的。  
类似地，函数 $\sin \frac{1}{x}$ 在 $x = 0$ 时并无意义的事实并不影响我们提出当 $x \to 0$ 它的极限的存在问题；但现在极限并不存在。

##### 55、极限理论的拓广
- 考察某一以 a 为聚点的区域 X 内定义的函数 $f(x)$ 。
- - $1^。$ 若当 x 趋于 a 时函数 $f(x)$ 有一有限的极限 A ，而 $A > p(A < q)$ ，则当 x 的值充分接近于 a （异于 a）时函数本身也就满足不等式
$$f(x) > p (f(x) < q) \ \ \ (14)$$  
选取正数 $ε < A - p(q - A)$ ，就有 $A - ε > p(A + ε < q)$ 。但依极限的定义，对于这 ε 必能求出 δ，只需 $|x - a| < δ$ （其中 x 取自 X 内且异于 a），就有 $A - ε < f(x) < A + ε$ 。对于这些 x 值，当然（14）式也能成立。
- - $2^。$ 若在 $x \to a$ 时函数 $f(x)$ 有一有限的正（负）极限，则函数本身也必为正（负）。至少，对于充分接近于 a 但异于 a 的 x 的数值是如此。
- - $4^。$ 若在 $x \to a$ 时函数 $f(x)$ 有一有限的极限，则对于充分接近于 a 的 x 的数值函数是有界的：$|f(x)| \le M'\ （M' = 常数，|x - a| < δ）$ 。  
有一有限极限的整序变量 $x_n$ ，最初也仅在 $n > N$ 时才能得出不等式 $|x_n| \le M'$ ，但因整序变量的数值仅只有限个数不能满足这不等式，故在有需要的时候也不难增大 M' ，使一切 $x_n$ 都能满足这不等式。但在函数的情形，一般说这是办不到的，因为使 $|f(x)| > M'$ 的 x 的数值可以是无穷集。例如，函数 $f(x) = \frac{1}{x}（x > 0）$ 在 $x \to 1$ 时，趋于 1 ；显然，若 $|x - 1| < \frac{1}{2}$ ，则 $f(x) < 2$ ，然而对于 x 的一切被考察的值，函数 $f(x)$ 缺不是有界的。
- 设在区域 X （有聚点 a）内给定两函数 $f(x)$ 及 $g(x)$ ；在 x 趋于 a 时有有限极限 $\lim f(x) = A ，\lim g(x) = B$ 。则函数 $f(x) \pm g(x) ，f(x) \cdot g(x) ，\frac{f(x)}{g(x)} \ \ \ (15)$ 也有有限极限（在商的情形须假定 $B \ne 0$），就是 $A \pm B ，A \cdot B ，\frac{A}{B}$ 。  
用“序列的语言”译述上述的两个关系，就成为：若 {$x_n$} 是 X 内的任意序列，有极限为 a ，则 $f(x_n) \to A ，g(x_n) \to B$ ，则立即得出 $\lim [f(x_n) \pm g(x_n)] = A ，\lim f(x_n)g(x_n) = A \cdot B ，\lim \frac{f(x_n)}{g(x_n)} = \frac{A}{B}$ 。
- 在 **31** 内所讲的关于“不定式”也都同样地自动转移到现在所考察的一般情形来了。它们用约定的记号
$$\frac{0}{0} ，\frac{\infty}{\infty} ，0 \cdot \infty ，\infty - \infty$$
来表示。如同在自然数变量函数的简单情形一样，这里为了要“确定不定式”，仅知道函数 $f(x)$ 及 $g(x)$ 的极限是不够的，必须计及这些函数的变动的规律。

##### 56、例题
- 1) 把 **32** 的例 1) 及 2) 普遍化，研究多项式
$$p(x) = a_0x^k + a_1x^{k-1} + ... + a_{k-1}x + a_k$$
而后再研究两个多项式的商
$$\frac{p(x)}{q(x)} = \frac{a_0x^k + a_1x^{k-1} + ... + a_{k-1}x + a_k}{b_0x^l + b_1x^{l-1} + ... + b_{l-1}x + b_l}$$
在 $x \to \pm \infty$ 的性态。经变形
$$p(x) = x^k(a_0 + \frac{a_1}{x} + ... + \frac{a_k}{x^k})$$
很容易确定 
$$\lim\limits_{x \to \pm \infty} p(x) = \pm \infty （\infty - \infty）$$
并且极限的符号可以这样确定：当 k 为偶数时它仅由 $a_0$ 的符号来决定，当 k 为奇数时除 $a_0$ 以外还要看 x 的符号。
- 2) 类似于此，可以求出 $\lim\limits_{x \to \pm \infty} \frac{p(x)}{q(x)} = \pm \infty ，\frac{a_0}{b_0} ，0 (\frac{\infty}{\infty})$ 。  
按照 $k > l ，k = l ，k < l$ 。极限的符号（在第二种情形）依 $a_0$ 及 $b_0$ 的符号而定，但在 $k - l$ 为奇数时还须依 x 的符号而确定。
- 3) 今将证明对于任意的正有理指数 r 的公式 $\lim\limits_{x \to 0} \frac{(1 + x)^r - 1}{x} = r （\frac{0}{0}）$。  
从指数为自然数：$r = n$ 这种最简单的情形开始。依牛顿二项定理
$$\frac{(1 + x)^n - 1}{x} = \frac{nx + \frac{n(n-1)}{1 \cdot 2}x^2 + ... + x^n}{x} = n + \frac{n(n-1)}{1 \cdot 2}x + ... + x^{n-1}$$
因为当 $x \to 0$ 时最后一式的一切项，除第一项外，都趋于 0 ，因此实际上就成为 $\lim\limits_{x \to 0} \frac{(1 + x)^n - 1}{x} = n$ 。  
今令 $r = \frac{1}{m}（m 是自然数）$ ，并考察式子 $\frac{\sqrt[m]{1+x} - 1}{x}$ 。假定 $\sqrt[m]{1+x} - 1 = y$ ，由此 $x = (1 + y)^m - 1$ 。  
因为（当作 $|x| < 1$）$1 - |x| < \sqrt[m]{1+x} < 1 + |x|$ 。所以 $\lim\limits_{x \to 0} \sqrt[m]{1+x} = 1$ 。  
于是，随同着 $x \to 0$ 已有 $y \to 0$ 。则依前面的情形，$\lim\limits_{x \to 0} \frac{\sqrt[m]{1+x} - 1}{x} = \lim\limits_{y \to 0} \frac{y}{(1 + y)^m - 1} = \frac{1}{m}$。  
最后，要证明一般的情形 $r = \frac{n}{m}$ ，仍引用辅助变数 y :
$$\frac{(1+x)^{\frac{n}{m}} - 1}{x} = \frac{(1 + y)^n - 1}{(1 + y)^m - 1} = \frac{(1 + y)^n - 1}{y} \cdot \frac{y}{(1 + y)^m - 1}$$
由此 $\lim\limits_{x \to 0} \frac{(1+x)^{\frac{n}{m}} - 1}{x} = \frac{n}{m}$ 。
- 4) 求极限 $\lim\limits_{x \to 0} = \frac{\sqrt[m]{1+x} - 1 - \frac{x}{m}}{x^2}$ 。  
仍用同样的代换式 $\sqrt[m]{1+x} - 1 = y$ ，则被考察的式子变为
$$\frac{y - \frac{1}{m}[(1 + y)^m - 1]}{[(1 + y)^m - 1]^2} = \frac{-\frac{m - 1}{2}y^2 + ...}{m^2y^2 + ...} = \frac{-\frac{m - 1}{2} + ...}{m^2 + ...}$$
由此，所求的极限等于 $-\frac{m - 1}{2m^2}$ 。
- 5) 极限 $\lim\limits_{x \to 0} \frac{\sin x}{x} = 1$ 经常被应用着去求其它极限。
 - a) $\lim\limits_{x \to 0} \frac{1 - \cos x}{x^2} = \frac{1}{2}（\frac{0}{0}）$  
 显然 $\frac{1 - \cos x}{x^2} = \frac{2\sin^2 \frac{x}{2}}{x^2} = \frac{1}{2}(\frac{\sin \frac{x}{2}}{\frac{x}{2}})^2 = \frac{1}{2}$ 。
 - b) $\lim\limits_{x \to 0} \frac{\tan x - \sin x}{x^3} = \frac{1}{2}（\frac{0}{0}）$  
 这里运用变形法很容易导向上面讨论过的极限：
 $$\frac{\tan x - \sin x}{x^3} = \frac{1}{\cos x} \cdot \frac{\sin x}{x} \cdot \frac{1 - \cos x}{x^2} = \frac{1}{2}$$
 - c) $\lim\limits_{x \to \frac{π}{2}} (\sec x - \tan x) = 0（\infty - \infty）$  
 在这里换成变量 $a = \frac{π}{2} - x$ 将更为便利；显然，当 $x \to \frac{π}{2}$ 时 $a \to 0$ 。我们就有
 $$\sec x - \tan x = \frac{1 - \sin x}{\cos x} = \frac{1 - \cos a}{\sin a} = \frac{1 - \cos a}{a^2} \cdot \frac{a}{\sin a} \cdot a \to 0$$

##### 57、单调函数的极限
- 设函数 $f(x)$ 定义在某一区域 $X = \{x\}$ 内。若对于这区域内的任意一对数值 x 与 x' ，由 $x' > x$ 能推得 $f(x') > f(x)[f(x') < f(x)]$ ，则 $f(x)$ 称为在区域内的**增函数（减函数）**。又若由 $x' > x$ 只能推得 $f(x') \ge f(x)[f(x') \le f(x)]$ ，则 $f(x)$ 称为**不减函数（不增函数）**。在这种场合，有时也称 $f(x)$ 是**广义增函数（减函数）**，一切这种类型的函数总称为**单调函数**。
- **定理** 设函数 $f(x)$ 在区域 X 内单调地增大，即使是广义的也可以。区域 X 以不小于一切 x 值的数 a（它可以是有限的数，或等于 $+\infty$）作为聚点。若在这时函数上有界： $f(x) \le M （对于 X 内的一切 x）$ ，则当 $x \to a$ 时函数有一有限极限；在与此相反的场合，它趋向于 $+\infty$ 。
- **证明** 假定函数 $f(x)$ 上有界，即当 x 在区域 X 内变动时函数值所成的集 {$f(x)$} 是上有界的。则这数集必有一有限的上确界 A 存在。今将证明这数 A 就是所求的极限。  
给定任意数 $ε > 0$ ，依上确界的性质，必能求出数值 $x' < a$ ，使 $f(x') > A - ε$ 。由于函数的单调性，则当 $x > x'$ 时更有：$f(x) > A - ε$ 。另一方面，因为永远有 $f(x) \le A \le A + ε$ ，故对满足上诉条件的 x 显然成立不等式 $|f(x) - A| < ε$ 。这就证明了本命题，只需在有限 a 时假定 $x' = a - δ$ ，而在 $a = +\infty$ 时取 $\Delta = x'$ 。  
若函数 $f(x)$ 不上有界，则不论 E 是怎样的数：必能求出 $x'$ 使 $f(x') > E$ ；而当 $x > x'$ 时当然更有 $f(x) > E$ 了。

##### 58、布尔查诺-柯西的一般判定法
- **定理** 函数 $f(x)$ 当 x 趋向于 a 时有一有限极限的必要且充分条件是，对于任一数 $ε > 0$ 必存在着 $δ > 0（\Delta > 0）$ 。只需
$$|x - a| < δ ，|x' - a| < δ （x > \Delta ，x' > \Delta）$$
就能成立不等式 $|f(x) - f(x')| < ε$ 。
- **证明** 在 a 是有限数的假定下证明
- **必要性** 设存在着有限极限 $\lim\limits_{x \to a} f(x) = A$ 。则依给定的 $ε > 0$ 必能求出 $δ > 0$ ，只需 $|x - a| < δ$ ，就有 $|f(x) - A| < \frac{ε}{2}$ 。又设 $|x' - a| < δ$ ，又有 $|A - f(x')| < \frac{ε}{2}$ 。由此，假定同时有 $|x - a| < δ$ 及 $|x' - a| < δ$ ，就得
$$|f(x) - f(x')| = |[f(x) - A] + [A - f(x')]| \le |f(x) - A| + |A - f(x')| < ε$$
- **充分性** 设在定理内所叙述的条件已经满足，又依任意取的 $ε > 0$ 能确定对应的 $δ > 0$ 。若 {$x_n$} 是 X 中 x 的任意收敛于 a 的序列，则依序列的极限的定义，必能求出序号 N ，当 $n > N$ 时能使 $|x_n - a| < δ$ 。与 n 同时又取另一序号 $n' > N$ ，于是同时有 $|x_n - a| < δ$ 及 $|x'_{n'} - a| < δ$ 。  
根据 δ 的选法，应有 $|f(x_n) - f(x'_{n'})| < ε$ 。因此，只要序号 n 及 n' 同时 > N ，这不等式就能成立，因而序列 $f(x_1), f(x_2), ..., f(x), ...$ 有一有限极限。  
由此已足证明不论怎样选取收敛于 a 的序列 {$x_n$} ，序列 {$f(x_n)$} 总趋于同一极限，这极限也就是函数的极限。

##### 59、函数的上极限及下极限
- 当 x 趋近于 a 时，即使函数 $f(x)$ 并无确定的极限存在，但是对于特定的序列 $x_n \to a$ 极限 $\lim\limits_{n \to +\infty} f(x_n)$ 仍可以存在，把它称为函数的**部分极限**。例如：$\sin x$ 在 $x \to \pm\infty$ 时。
- 在函数的部分极限中恒能求出最大的与最小的，称为它的上极限与下极限，并记成 $\overline{\lim}\limits_{x \to a}f(x)$ 及 $\underline{\lim}\limits_{x \to a}f(x)$ 。
- 函数有确定极限（依通常的意义）存在的必要且充分的条件是：上极限与下极限相等。

#### 三、无穷小及无穷大的分阶
##### 60.无穷小的比较
假定在作某种研究时须同时考察一系列的无穷小量：α，β，γ 。一般地说来，他们是同一变量 x 的函数。而 x 是趋于有限极限或无穷极限 a 的变量。二无穷小 α 及 β 的比较以其比式的性态为基础。
- Ⅰ 若比式 $\frac{β}{α}（\frac{α}{β}亦一样）$ 有一异于零的有限极限，则无穷小 α 与 β 称为是同阶的。
- Ⅱ 若比式 $\frac{β}{α}$ 趋于无穷小（相反地，比式 $\frac{α}{β}$ 趋于 $\infty$），则无穷小 β 称为是比 α 高阶的无穷小，同时无穷小 α 就是比 β 低阶的无穷小。
- 例如，若 $α = x \to 0$ ，则 $\sin x ，\tan x ，\sqrt[m]{1 + x} - 1$ 与这一无穷小相比较都是与它同阶的无穷小，因为：$\lim\limits_{x \to 0} \frac{\sin x}{x} = 1 ，\lim\limits_{x \to 0} \frac{\sqrt[m]{1 + x} - 1}{x} = \frac{1}{m}$ 。反之，无穷小
$$\sqrt[m]{1 + x} - 1 - \frac{x}{m} ，1 - \cos x ，\tan x - \sin x \ \ \ (1)$$
显然是比 x 高阶的无穷小。
- 当然也可能有二无穷小的比式并不趋向于任何极限的情形。
- 无穷小 β 是比无穷小 α 更高阶的，则这一事实就写成：$β = o(α)$ 。例如：$1 - \cos x = o(x) ，\tan x - \sin x = o(x)$ 。

##### 61.无穷小的尺度
有时为了要更精确地比较无穷小的性态，需要用数字来表达它们的阶。在这种情形，首先在所研究的无穷小内选出一个（就说是 α）作为一种“标准”，把它称为**基本无穷小**。基本无穷小的选取在某种程度内是任意的，但通常总选取最简单的。例如，假定被考察的量都是 x 的函数，而它们当 x 趋向于 a 时成为无穷小，那么根据 a 是零，是异于零的有限数或是无穷大，依次选取 $x，x - a，\frac{1}{x}$ 作为基本无穷小。  
再由基本无穷小 α（α > 0）的各种正指数幂 $α^k$ 来组成一尺度，去评较性质上更为复杂的无穷小。
- Ⅲ 若 β 与 $α^k$ 是同阶的无穷小量，即若比式 $\frac{β}{α^k}$ 是有异于零的有限极限，则称无穷小 β 为关于基本无穷小 α 的 k 阶无穷小量。
- 例如（1）中前面两个是关于 $α = x$ 的**二阶**无穷小，而最后一个是**三阶**无穷小，因为：
$$\lim\limits_{x \to 0} \frac{\sqrt[m]{1 + x} - 1 - \frac{x}{m}}{x^2} = -\frac{m - 1}{2m^2} ，\lim\limits_{x \to 0} \frac{1 - \cos x}{x^2} = \frac{1}{2} ，\lim\limits_{x \to 0} \frac{\tan x - \sin x}{x^3} = \frac{1}{2}$$
- 试考察表达式 $β = \sqrt{x + 1} + \sqrt{x - 1} - 2\sqrt{x}$ 。  
在 $x \to \infty$ 时它将是无穷小：
$$β = (\sqrt{x + 1} - \sqrt{x}) - (\sqrt{x} - \sqrt{x - 1}) = \frac{1}{\sqrt{x + 1} + \sqrt{x}} - \frac{1}{\sqrt{x} + \sqrt{x - 1}}$$
继续这变形：
$$β = \frac{\sqrt{x - 1} - \sqrt{x + 1}}{(\sqrt{x + 1} + \sqrt{x})(\sqrt{x} + \sqrt{x - 1})} = \frac{-2}{(\sqrt{x + 1} + \sqrt{x})(\sqrt{x} + \sqrt{x - 1})(\sqrt{x - 1} + \sqrt{x + 1})}$$
今令 $α = \frac{1}{x}$ 
$$\lim\limits_{x \to +\infty} \frac{β}{α^{\frac{3}{2}}} = \frac{-2(\sqrt{x})^3}{(\sqrt{x + 1} + \sqrt{x})(\sqrt{x} + \sqrt{x - 1})(\sqrt{x - 1} + \sqrt{x + 1})}$$
$$= \frac{-2}{(\sqrt{1 + \frac{1}{x}} + \sqrt{\frac{1}{x}})(\sqrt{\frac{1}{x}} + \sqrt{1 - \frac{1}{x}})(\sqrt{1 - \frac{1}{x}} + \sqrt{1 + \frac{1}{x}})}$$
$$= -\frac{1}{4}$$
这样，β 的阶就是 $\frac{3}{2}$ 。  
不是任一无穷小 β （即使是与一切幂 $α^k$ 都能比较的）都有确定的阶数。  
$$\lim\limits_{x \to +\infty} \frac{a^x}{x^k} = +\infty ，\lim\limits_{x \to +\infty} \frac{\log_a x}{x^k} = 0$$
由此：
$$\lim\limits_{x \to +\infty} \frac{x^k}{a^x} = 0 ，\lim\limits_{x \to +\infty} \frac{x^k}{\log_a x} = \infty$$
现在再把式中的 x 换成 $\frac{1}{x}$ ，并在第一式内假定 $a = \frac{1}{c}，0 < c < 1$ ，则得：
$$\lim\limits_{x \to +0} \frac{c^{\frac{1}{k}}}{x^k} = 0 ，\lim\limits_{x \to +0} \frac{\frac{1}{\log_a x}}{x^k} = \infty$$
这样，$c^{\frac{1}{k}}（0 < c < 1）$ 就成为比一切幂 $x^k（k > 0）$ 更高阶的无穷小，而同时 $\frac{1}{\log_a x}（a > 1）$ 成为比一切幂 $x^k$ 更低阶的无穷小。

##### 62、等价无穷小
- Ⅳ 无穷小 α 与 β 称为等价无穷小（记为**α~β**），若它们的差 $γ = β - α$ 是比 α 及 β 中的任何一个更高阶的无穷小：$γ = o(α)$ 及 $γ = o(β)$ 。  
只要 γ 是比这些无穷小之一更高阶就已够了，因为，例如，若 γ 是比 α 更高阶，则它亦必比 β 更高阶。事实上，由 $\lim \frac{γ}{α} = 0$ ，就能推得：$\lim \frac{γ}{β} = \lim \frac{γ}{α + γ} = \lim \frac{\frac{γ}{α}}{1 + \frac{γ}{α}} = 0$ 。
- 考察两个等价无穷小 α 及 β ，设 $β = α + γ$ ，其中 $γ = o(α)$ 。若近似地假定 $β = α$ ，则当他们的值都在减小时，不但由这代替所生的绝对误差 $|γ|$ 趋于零，而且相对误差 $|\frac{γ}{α}|$ 也趋于零。近似等式 $β = α$ 在 α 及 β 的数值充分小时可以有任意大的准确度。据此，在近似运算内，繁复的无穷小可以换成与它等价的的简单无穷小。
- 建立一个有用的检定二无穷小的等价性的方法，在本质上，它就给出这概念的第二定义，与前面所给的定义等价：**要使二无穷小 α 与 β 成为等价的，其充要条件为 $\lim \frac{β}{α} = 1$**。  
充分性。先设这关系成立，于是 $δ = \frac{β}{α} - 1 \to 0$ 。则 $γ = β - α = δ \cdot α$ 就是比 α 更高阶的无穷小，因为 $\lim \frac{γ}{α} = \lim δ = 0$ 。  
必要性。先设 α 与 β 是等价的，即 $γ = β - α$ 是比 α 更高阶的无穷小。由此就有 $\frac{β}{α} - 1 = \frac{γ}{α} \to 0$ ，故 $\frac{β}{α} \to 1$ 。
- 用这检定法，立即看出，在 $x \to 0$ 时无穷小 $\sin x$ 及 $\tan x$ 是与 x 等价的，而 $\sqrt[m]{1 + x} - 1$ 是与 $\frac{1}{m}x$ 等价的。由此就得出近似公式：$\sin x = x ，\tan x  = x ，\sqrt[m]{1 + x} - 1 = \frac{1}{m}x ，特例，\sqrt{1 + x} - 1 = \frac{1}{2}x$ 。
- 等价无穷小的已证明的性质可以应用于确定形如 $\frac{0}{0}$ 的不定式。即确定二无穷小的比式 $\frac{β}{α}$ 的极限。这时，它们中的任一个可以换成任何与它等价的无穷小，而对于极限的存在及数值并无影响。  
事实上。若 $\overline{α} \sim α$ ，$\overline{β} \sim β$ ，即 $\lim \frac{\overline{α}}{α} = 1$ ，又 $\lim \frac{\overline{β}}{β} = 1$ ，则比式 $\frac{β}{α} = \frac{β}{\overline{β}} \cdot \frac{\overline{β}}{\overline{α}} \cdot \frac{\overline{α}}{α}$ ，与比式 $\frac{\overline{β}}{\overline{α}}$ 的区别仅是多了两个趋于 1 的因式，因此与它同时趋于同一极限。
- 若能把 α 及 β 选取得足够简单，则立即可使问题大为简化，例如：$\lim\limits_{x \to 0} \frac{\sqrt{1 + x + x^2} - 1}{\sin 2x} = \lim\limits_{x \to 0} = \frac{\frac{1}{2}(x + x^2)}{2x} = \frac{1}{4}$
- 都与第三者等价的二无穷小是等价的。

##### 63、主部的分出
若选定 α 作为基本无穷小，则形如 $c \cdot α^k$ 的量自然就认为是**最简单**的无穷小，此处 c 是常系数，而 $k > 0$ 。设 β 是关于 α 的 k 阶无穷小，即 $\lim \frac{β}{α^k} = c$ ，式中 c 是异于零的有限数，则 $\lim \frac{β}{cα^k} = 1$ ，而无穷小 β 与 $cα^k$ 就是等价的无穷小： $β \sim cα^k$ 。
- 与给定的无穷小 β 等价的这个最简单的无穷小 $cα^k$ 就称为 β 的**主部**（或主项）。
- 很易分出下列各式的主部：$1 - \cos x \sim \frac{1}{2}x^2 ，\tan x - \sin x \sim \frac{1}{2}x^3$ ，此处 $x \to 0$ 。而 $α = x$ 本身就是基本无穷小。  
若 $x \to +\infty$ 而采用 $α = \frac{1}{x}$ 作为基本无穷小，就有 $\sqrt{x + 1} + \sqrt{x - 1} - 2\sqrt{x} \sim -\frac{1}{4}(\frac{1}{x})^{\frac{3}{2}}$
- 设 $β \sim cα^k$ 。即 $β = cα^k + γ$ ，式中 $γ = o(α^k)$ 。可以想象，从无穷小 γ 内再可以分出主部：$γ = c'α^{k'} + δ$ ，式中 $k' > k$ 。而 $δ = o(α^{k'})$ 。  
例如，假定 $x \to 0$ ：$\sqrt[m]{1 + x} - 1 = \frac{1}{m}x + γ$ ，$\lim\limits_{x \to 0} \frac{γ}{x^2} = -\frac{m - 1}{2m^2}$ ，于是 γ 的主部就是 $-\frac{m - 1}{2m^2}x^2$ 。由此 $\sqrt[m]{1 + x} - 1 = \frac{1}{m}x -\frac{m - 1}{2m^2}x^2 + o(x^2)$ ，特别是 $\sqrt{1 + x} - 1 = \frac{1}{2}x -\frac{1}{8}x^2 + o(x^2)$ 。  
这种从无穷小内逐次分出阶数始终在增高的最简单的无穷小的步骤，可以继续进行下去。
- 若已知二无穷小 β 及 γ 的主部是 $cα^k$ 及 $c'α^{k'}$ ，则关于其和 $β + γ$ 的主部。  
在 $k \ne k'$ 时，它的主部显然就是 $cα^k$ 及 $c'α^{k'}$ 两项中指数较小的那一项。今设 $k = k'$ ，则 $β + γ$ 的主部就是 $(c + c')α^k$ ，唯一需要假定 $c + c' \ne 0$ 。但当两个主部互相对消的情形，则和 $β + γ$ 将是比任一加数更高阶的无穷小。  
例如，在 $x \to 0$ 时，对于无穷小 $β = \sqrt{1 + x} - 1 \sim \frac{1}{2}x$ 及 $\sqrt{1 - x} - 1 \sim -\frac{1}{2}x$ 就是如此。若再分出它们以后的主部：$β = \frac{1}{2}x - \frac{1}{8}x^2 + o(x^2)$ ，$γ = -\frac{1}{2}x - \frac{1}{8}x^2 + o(x^2)$ ，则很明显有 $β + γ = \sqrt{1 + x} + \sqrt{1 - x} - 2 = -\frac{1}{4}x^2 + o(x^2)$ ，于是 $β + γ$ 就是二阶无穷小，而它的主部等于 $-\frac{1}{4}x^2$ 。

- 1) 设用长 l 米的尺测量某一地方的直线距离。因为实际上没有把尺准确地沿着测量的直线放置，以致测量的结果显得比真实的长度大了一些。试就最坏的情形而论，假设在测量时把尺放成锯齿形，就是说，它的两端交迭地忽而偏在直线的一侧忽而偏在另一侧，其离开直线的距离为 λ 米（图25）。今试估计其误差。![25.jpg](./images/25.jpg)  
在尺每放下一次时所发生的绝对误差等于尺的长度 l 与它在所测量的直线上的投影的差，其投影是：
$$2\sqrt{(\frac{l}{2})^2 - λ^2} = l\sqrt{1 - \frac{4λ^2}{l^2}}$$
应用近似公式：$\sqrt{1 + x} = \frac{1}{2}x + 1$ 于 $x = - \frac{4λ^2}{l^2}$ 的场合（由于 λ 比 l 小得多，所以这样做是可以的），可以把投影的长度换成下式：
$$l(1 - \frac{2λ^2}{l^2}) = l - \frac{2λ^2}{l}$$
因此，绝对误差是 $\frac{2λ^2}{l}$ ，而相对误差显然是 $\frac{2λ^2}{l^2}$ 。这相对误差并不随所量距离的长短而改变。  
若限制相对误差不能大于 δ ，即应有 $\frac{2λ^2}{l^2} < δ$ ，则必须 $λ < l\sqrt{\frac{δ}{2}}$ 。  
例如，在用 2 米的尺（$l = 2$）测量时，要达到 0.001 的相对准确度，只要偏差 λ 不超过 $2\sqrt{0.0005} = 0.045 米 = 4.5 厘米$ 就足够了。

这样 $l = π(R + r) + 2a(R -r) + 2\sqrt{d^2 - (R -r)^2}$ 。

把这些数值代入并整理化简就得出最后的公式：$l = π(R + r) + 2d + \frac{(R - r)^2}{d}$ 。

##### 65、无穷大的分阶
对于无穷大量也可以进行相似的分阶，将所考察的诸无穷大量当作是同一变量 x 的函数，当 x 趋于 a 时它们趋于 $\infty$ 。
- Ⅰ 二无穷大 y 及 z 称为是同阶的无穷大，若它们的比式 $\frac{y}{z}（\frac{z}{y}也如此）$ 有异于零的有限极限。
- Ⅱ 但若比式 $\frac{z}{y}$ 趋于 $\infty$ （而比式 $\frac{y}{z}$ 趋于零），则称 z 为比 y 更高阶的无穷大，而同时 y 称为比 z 更低阶的无穷大。
- 当比式 $\frac{z}{y}$ 不趋于任何极限时。无穷大 y 与 z 就是不能比较的。
- 在同时考察一系列的无穷大量时，可选取其中之一当作基本无穷大，而其余的无穷大就与它的幂相比较。若它们全部都是 x 的函数，且当 $x \to a$ 时趋于 $\infty$ ，那么，通常若 $a = \pm \infty$ 就取 $|x|$ 本身，因而 a 为有限数时就取 $\frac{1}{|x - a|}$ 作为基本无穷大。
- Ⅲ 无穷大 z 称为 k 阶的无穷大（关于基本无穷大 y），如果 z 与 $y^k$ 是同阶的话，即若比式 $\frac{z}{y^k}$ 有异于零的有限极限。
- 当 $x \to +\infty$ 时 $a^x （a > 1）$ 是比任何幂 $x^k$ 更高阶的无穷大（k 是正指数），而 $\log_a x （a > 1）$ 是比任何 $x^k$ 更低阶的无穷大，这是从 [**61** 公式(2)] 推得的。

#### 四、函数的连续性及间断性
##### 66、函数在一点处的连续性的定义
- 考察定义在以 $x_0$ 为聚点的某个区域 $X = \{x\}$ 内的函数 $f(x)$ ；并设点 $x_0$ 本身属于 X ，于是在这点函数有确定的数值 $f(x_0)$ 。当建立函数在 x 趋于 $x_0$ 时的极限概念 $\lim\limits_{x \to x_0} f(x)$ 时，曾不止一次地着重指出，变数 x 并不取数值 $x_0$ ；这数值甚至可以不属于函数的定义域，即使它属于这定义域，但在研究上述极限时数值 $f(x_0)$ 也不考虑在内。
- 然而，正是当 $\lim\limits_{x \to x_0} f(x) = f(x_0) \ \ \ (1)$ 的情形才有着特殊的重要性。常说，函数 $f(x)$ 当 $x = x_0$ 时（或在点 $x = x_0$ 处）是连续的，只要关系式（1）能成立；又若它不成立，就说当 x 取这数值（或在这点）函数有间断。
- 当函数 $f(x)$ 在点 $x_0$ 处为连续的场合（只限于这种场合）若要计算 $f(x)$ 当 $x \to x_0$ 时的极限，就可以不管 x 在本身趋向于 $x_0$ 的过程中是否特殊地取过数值 $x_0$ 。
- 函数的连续性的定义也可以用其他的术语来叙述。由数值 $x_0$ 过渡到旁的数值 x 。可以想象为给数值 $x_0$ 加上一个增量 $\Delta x_0 = x - x_0$ 。函数的新值 $y = f(x) = f(x_0 + \Delta x_0)$ 与原值 $y_0 = f(x_0)$ 相差一个增量 $\Delta y_0 = f(x) - f(x_0) = f(x_0 + \Delta x_0) - f(x_0)$ 。  
要使函数 $f(x)$ 在点 $x_0$ 处是连续的，其充要条件是：在这点函数的增量 $\Delta y$ 与自变数的增量 $\Delta x_0$ 一同趋向于零。换句话说：连续函数的特性就是，对应于变元的无穷小增量，函数的增量也是无穷小。
- 函数 $f(x)$ 在点 $x_0$ 处的连续性的意义可以归结如下：对于不论怎样的数 $ε > 0$ 必能求出数 $δ > 0$ ，使由 $|x - x_0| < δ$ 可以引出 $|f(x) - f(x_0)| < ε$ 。这样，最后的不等式在点 $x_0$ 的充分小的邻域 $(x_0 - δ，x_0 + δ)$ 内就应该成立。
- 用“序列的语言”表达连续性：在 X 内任意取收敛于 $x_0$ 的 x 的序列：$x_1, x_2, ..., x_n, ...$ 则对应的函数值的序列 $f(x_1), f(x_2), ...., f(x_n), ...$ 必收敛于 $f(x_0)$ 。
- **附注** 设点 $x = x_0$ 是函数 $f(x)$ 的定义区域 X 的聚点，但本身不属于 X ，于是函数在这点是没有定义的。然若存在着有限极限 $\lim\limits_{x \to x_0} f(x)$ ，那么只要补充函数的定义，令 $f(x_0)$ 等于这极限，$f(x)$ 就在 $x = x_0$ 处是连续的了。  
反之，若所说的极限并不存在，那么即使函数在这一点 $x = x_0$ 没有确定也好，总是说，函数在这一点遭受间断：这时不论函数在 $x = x_0$ 补取什么数值，它在此处还是有间断。
- 要考察在区间 X 内确定的函数，区间中的一切点都是它的聚点，于是对于其中的任何一点都可以提出有关连续性的问题。若函数在区间 X 内的每一点都是连续的，就说函数在区间 X 内是连续的。
