### 绪论 实数
#### 一、有理数域
##### 1、前言
- 有理数：整数和分数统称有理数，任何一个有理数都可以写成分数$\frac {m}{n}$（m,n都是整数,且n≠0）的形式。
- 在有理数范围，并非一切线段都有长度。例：单位长度的正方形，其对角线的长度，根据勾股定理为 $\sqrt{2}$，显然这是不可能的。

##### 2、有理数的序
- Ⅰ1  每一对数a与b之间必有且仅有下列关系之一： a = b, a > b, a < b；
- Ⅰ2  由 a > b 且 b > c 推得：a > c（>的传递性）；
- Ⅰ3  由 a > b，则必能求得一数 c，使 a > c， 且 c > b（稠密性）。

##### 3、有理数的加法及减法
- Ⅱ1  a + b = b + a （加法的交换性）；
- Ⅱ2  (a + b) + c = a + (b + c) （加法的结合性）；
- Ⅱ3  a + 0 = a；
- Ⅱ4  对于每一数 a 存在着（与它对称的）数 -a，使得 a + (-a) = 0；
- Ⅱ5  由 a > b 推得 a + c > b + c。

##### 4、有理数的乘法及除法
- Ⅲ1 ab = ba （乘法的交换性）；
- Ⅲ2 (ab)c = a(bc) (乘法的结合性）；
- Ⅲ3 a·1 = a ；
- Ⅲ4 对于每一个不等于0的数a，比有数 $\frac {1}{a}，使得 a·\frac {1}{a} = 1$ ；
- Ⅲ5 （a + b）c = a·c + b·c （乘法关于和的分配性）；
- Ⅲ6 a > b 及 c > 0 推得 a·c > b·c ；

##### 5、阿基米德公理
- Ⅳ1 不论 c > 0 是怎样的数，总有大于 c 的自然数 n 存在（阿基米德公理）

#### 二、无理数的导入，实数域的序
##### 6、无理数的定义
- 若将有理数全体所成的集合分拆为两个非空集合，这样的分拆叫做分划。
- 1 任一有理数，必在且仅在 A 及 A' 二集之一中出现。
- 2 集 A 内的任一数 a ， 必小于集 A'内的任一数 a'。
- 集 A 称为分划下组，集 A' 为上组，分划记成 A|A'。
- 小于下组内的数 a 的一切有理数也都属于下组，大于上组内的数 a' 的一切有理数亦都属于上组。
- 分划仅能有三种类型：
 - 1 在下组 A 内无最大数，而在上组 A' 内有最小数 r；
 - 2 在下组 A 内有最大数r，而在上组 A' 内无最小数；
 - 3 在下组内既无最大数，而在上组内亦无最小数。
- 前两张情形，分划由有理数 r 所产生（界数）,或者说分划定义有理数 r 。
- 任意 3 型的分划定义某一无理数α，这个数α代替缺少的界数，把它插入在 A 组的一切数 a 与 A' 组的一切数 a' 中间。
- 有理数及无理数总称为实数。

##### 7、实数域的序
- Ⅰ1 任一对实数α与β之间必有且仅有下列三种关系之一： α = β， α > β， α < β；
- Ⅰ2 由α > β，β > γ 推得 α > γ；

##### 8、辅助命题
- 引理1 对于不论怎样的两个实数α及β，其中α > β，恒有一个位于它们中间的有理数 r : α > r > β；
- 引理2 设给定两个实数α和β。如果任取一个数 e > 0，数α及β都能位于同一对有理数 s 与 s' 之间：s' > α > s，s' > β > s，这对数的差小于 e： s' - s < e ，则α与β必相等。

##### 9、用无限循环小数表示实数
- (1) $C_0.c_1c_2...c_n < α < C_0.c_1c_2...c_n + \frac {1}{10^n}$
- (1a) $C_0.c_1c_2...c_n <= α <= C_0.c_1c_2...c_n + \frac {1}{10^n}$

##### 10、实数域的连续性
- 把实数域分拆为两个均非空集 A 及 A'，使能满足：
 - $1^。$ 每一实数必落在 A，A' 中一个且仅一个之内；
 - $2^。$ 集 A 内的每一数 α 小于集 A' 内的每一数 α' 。
- <b>基本定理(戴德金)</b>：对于实数域内的任一分划 A|A' 必有产生这一分划的实数 β 存在。这数 β ，1) 或是下组 A 的最大数，2) 或是上组 A' 内的最小数。

实数的这一性质常称为它的<b>完备性</b>，也称为它的<b>连续性</b>。

##### 11、数集的界
- x 代表集内的一般数，诸数 x 所成的集，记为：<b> X = {x}</b>。
- 若有这样的数 M 存在，使一切 x <= M ，就说，这集被 M <b>上有界</b>，这 M 就是集 {x} 的<b>上界</b>。
- 若有这样的数 m 存在，使一切 x >= m ，就说，这集被 m <b>下有界</b>，这 m 就是集 {x} 的<b>下界</b>。
- 上(下)有界的集，也可以又上(下)有界，也可以不是上(下)有界。
- 若数集不上(下)有界，则称“广义的数”<b>+∞(-∞)</b>为它的上(下)界。
 - -∞ < +∞ 及 -∞ < α < +∞
- 上界内的最小数称为<b>上确界</b>，下界内的最大数称为<b>下确界</b>。
- <b>定理</b> 若集 X = {x} 上(下)有界，则它必有上(下)确界。

#### 三、实数的算术运算
##### 12、实数的和的定义
- (1) a < α < a' 及 b < β < b'
- (2) a + b < γ < a' + b'
- 则 γ 称为 α 及 β 的和，记为 α + β

##### 13、加法的性质
- Ⅱ1<sup>。</sup> α + β = β + α；
- Ⅱ2<sup>。</sup> (α + β) + γ = α + (β + γ)；
- Ⅱ3<sup>。</sup> α + 0 = α；
- Ⅱ4<sup>。</sup> 对于任意实数 α，存在着（对称于它的）数 -α，满足条件 α + (-α) = 0；
- Ⅱ5<sup>。</sup> 由 α > β 推得 α + γ > β + γ。

##### 14、实数的积的定义
- a < α < a'、b < β < b'；
- 位于一切形如 ab 的积与一切形如 a'b'的积之间的实数 γ:
 - (3) ab < γ < a'b'；
- 称为 α 及 β 的积，记为 αβ。
- 不论 α 是怎样的实数，常有：
 - α · 0 = 0 · α = 0
- 二数都异于 0 时，则根据通常的“符号规则”：
  - 当 α 与 β 同号时：α · β = |α| · |β|
  - 当 α 与 β 异号是：α · β = -(|α| · |β|)

##### 15、乘法的性质
- Ⅲ$1^。$ α · β = β · α；
- Ⅲ$2^。$ (α · β)γ = α(β · γ);
- Ⅲ$3^。$ α · 1 = α；
- Ⅲ$4^。 对于任一异于 0 的实数 α，必有倒数 \frac {1}{α} 存在，满足条件：α · \frac {1}{α} = 1$；
- Ⅲ$5^。$ (α + β) · γ = α · γ + β · γ；
- Ⅲ$6^。$ 由 α > β 及 γ > 0 推得 α · γ > β · γ。

##### 16、阿基米德公理
- Ⅳ$1^。$ 对不论怎样的实数 γ，必有大于 γ 的自然数 n 存在。

##### 17、绝对值
- |α| < β（β > 0）相当于 -β < α < β
- |α| <= β（β > 0）相当于 -β <= α <= β
- |α + β| < |α| + |β|
- |α - β| < |α| + |β|
- ||α| - |β|| < |α - β|

#### 四、实时的其他性质及应用
##### 18、根的存在、以有理数为指数的幂
- 设 α 是任一实数，n 是自然数，实数 ξ 称为数 α 的 n 次根： $ξ^n$ = α 。
- 限定 α 是正数，则满足于这一关系式的正数 ξ 就是<b>根的算术值</b>。
- $α^r  · α^{r'} = α^{r + r'} ，\frac {α^r}{α^{r'}} = α^{r - r'}$，
- $(α^r)^{r1} = α^{r · r1} ，(αβ)^r = α^r · β^r$ ，
- $(\frac {α}{β})^r$ = $\frac {α^r}{β^r}$ 。
- 在 α > 1 时，幂 $α^r$ 随指数 r 的增大而增大。

##### 19、以任一实数为指数的幂
- b 及 b' 为有理数，且满足不等式：b < β < b' ，位于所有 $α^b 及 α^{b'}之间的实数 γ：α^b < γ < α^{b'}$ 称为数 α > 1 的 β 次幂。
- <b>伯努利不等式</b>：如果 n 是大于 1 的自然是，且 γ > 1 ，则：$γ^n > 1 + n(γ - 1)$ 。
- 设 $γ = α^{\frac {1}{n}}(α > 1)，则得不等式：α^{\frac {1}{n}} - 1 < \frac {α - 1}{n}$
- $α^β · α^γ = α^{β + γ}$

##### 20、对数
- α > 1 ，且 γ 为正实数，满足：$α^r = γ$ ，则 r 为以 α 为底的 γ 的对数。
- $α^β = γ, β = log_αγ$。

##### 21、线段的度量
- 对于任一直线段 A 以一个正实数 l(A) 和它对应，即称为<b>线段A的长</b>使得：
 - 1) 某一预先指定的线段 E(长的单位)有长 1 : l(E) = 1;
 - 2) 相等的线段有同一的长；
 - 3) 在线段相加时，和的长等于各相加线段之长的和：l(A + B) = l(A) + l(B) 。
- 若在轴上选取原点 O 及 单位长 OE 则：
 - 在直线上的任一点 X 必对应于某一实数x，称为它的横标，若 X 位于 O 起的正向上，x 即等于 OX 线段的长，而在相反的情形等于这长的负数。
